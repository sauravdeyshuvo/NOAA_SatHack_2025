```python
import h5py

# Data 
fp = r'D:\1Research\2025\NOAA_SatHack\data\glm_ca_subset\OR_GLM-L2-LCFA_G16_s20202290000000_e20202290000200_c20202290000227.nc'

def print_h5_structure(name, obj):
    """
    Callback for h5py.File.visititems():
    Prints group/dataset names and their shapes.
    """
    if isinstance(obj, h5py.Dataset):
        print(f"Dataset: {name}  shape: {obj.shape}")
    else:
        print(f"Group:   {name}")

with h5py.File(fp, 'r') as h5:
    h5.visititems(print_h5_structure)
```

```python
from netCDF4 import Dataset

fp = r'D:\1Research\2025\NOAA_SatHack\data\glm_ca_subset\OR_GLM-L2-LCFA_G16_s20202290000000_e20202290000200_c20202290000227.nc'

with Dataset(fp, 'r') as ds:
    print(ds)                          # prints global attributes & dimensions
    print("Variables:")
    for var in ds.variables:
        print(" ", var,
              "dims:", ds.variables[var].dimensions,
              "shape:", ds.variables[var].shape)
```
